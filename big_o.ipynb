# Big O Notation

> O(f(N))


* **Upper bound** of an algorithm's time or space complexity
* Asymptotic: approaching some behavior at infinity
*   *N* = size, *f(N)* = function of algorithm operations




Helps in understanding the scalability of algorithms and predicting how they will perform as the input size grows.

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

# x-coordinates
x1 = np.arange(0, 2*(np.pi), 0.01)
# y-coordinates
y1 = np.cos(2.0*x1*np.pi)

# plot values as a figure
fig = plt.figure()
ax = fig.add_subplot(111)
ax.set_xlim(0, 1.0)
ax.set_ylim(-1.2, 1.2)
ax.plot(x1,y1)
```

## Definition
f(n) is O(g(n)) if there exists constants c **> 0** such that f(n) grows no faster than c\*g(n) for all n â‰¥ $n_0$
